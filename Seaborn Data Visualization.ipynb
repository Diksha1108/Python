```python
# joint plot....

# sns.jointplot(x = "sepal_width", y = "sepal_length", data = df, kind = "scatter")

sns.jointplot(x = "sepal_width", y = "sepal_length", data = df, kind = "reg")
```

```python
### pair plot
```

```python
sns.pairplot(df, hue = "petal_width")
```

```python
# dis plot


sns.distplot(df['sepal_width'],kde = False, bins = 5)
```

```python
# count plot

sns.countplot("petal_length", data = df)
sns.countplot("petal_width", data = df)
```

```python
# bar plot..

sns.barplot(y = "sepal_length", x = "sepal_width", data = df)
```

```python
# box plot

sns.boxplot("sepal_width","sepal_length", data = df)
```

```python
sns.boxplot(x = "petal_width", y = "petal_length",hue = "species",data = df)
```
